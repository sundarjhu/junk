# Basic ideas/terminology:

<font color='red'>**Randomness is unpredictability**</font>. It means that repeated observations of the same quantity will result in slightly differing results each time. It means that no matter how many observations you already have, <font color='red'>**you will never be able to predict the next observation with absolute certainty**</font>.

   One way to think about it is that Nature has some pre-determined formula for which set of values it will access to describe a given phenomenon. Each time you perform an observation, Nature "draws" a single value from this set and returns it to you. Upon repetition, a different value is returned. No matter how many times you repeat the experiment -- in other words, how many <font color='red'>**samples**</font> you draw --, you will never be able to discover the full set of values -- the <font color='red'>**population**</font> -- Nature is using. However, as the number of experiments increase, you will be able to obtain an increasingly good approximation of this set.

   The population as known only to Nature has certain properties that summarise its location, spread, etc. These are called <font color='red'>**population quantities or parameters**</font>. Examples include the <font color='red'>**true mean**</font> (an estimate of the location) and <font color='red'>**true standard deviation**</font> (an estimate for the spread) of the population distribution. We are only ever able to obtain a <font color='red'>**finite sample**</font> from this population, which we can nevertheless use this finite sample to estimate the population parameters. So, for example, we can compute the <font color='red'>**sample mean and sample standard deviation**</font> which will then act as estimates for the respective population quantities.
   
   Although individual observations are impossible to predict with certainty, you can get a better idea of <font color='red'>**long-term trends**</font> (e.g., the average number of heads obtained from 100 coin tosses) over repeated observations (i.e., by increasing your sample size).


# Illustration: a coin toss
>The coin toss is a very simple example to demonstrate the concepts above. Each coin flip will result in one of only two outcomes: a head (H) or a tail (T).
>
>No matter how many times we have already flipped the coin and recorded the result, we cannot know with certainty the outcome of the next coin flip; however, we can estimate the probability of the result being one of the two outcomes (H or T).
>
>For example, let's say you have already flipped the coin two times, resulting in two heads. The probability of the next flip returning heads is then 2/2 = 1. <font color='red'>**According to your sample**</font>, it is absolutely certain that the next flip will return heads!
>
>Only Nature knows the true probability of returning heads on a coin flip. What we have done above is to estimate this population parameter using the data we had. A 100% certainty seems suspect, but we have done the best we can with the data we have.


## Functions of random numbers are also random!
>In the above example, we tossed a coin twice and obtained two heads. We could repeat this sequence of experiments ("toss a coin twice") many times. Let's say we did just this, and obtained the following set of results:

>>First sequence: tossed coin twice, two heads.</br>
Second sequence: tossed coin twice, two heads.</br>
Third sequence: tossed coin twice, one head and one tail.</br>
Fourth sequence: tossed coin twice, two tails.</br>

>For each of the above sequences, we can also estimate the probability of obtaining heads on the next coin flip, the results being 100%, 100%, 50%, and 0% respectively. <font color='red'>**Each time the sample changes, the sample-based estimate of the true probability changes**</font>.

>This tells us that <font color='red'>**functions of random numbers are also random numbers**</font>. Just as with the random numbers themselves, these functions also have their own <font color='red'>**true (population) distribution**</font>, and just as with the random numbers, we can only ever estimate the true quantities associated with these functions because we will only have a finite number of <font color='red'>**realisations**</font> (sequences, in our example above).


# The Law of Large Numbers
> Let's go back to the example of two coin flips resulting in two heads. We estimated that the probability of the next flip resulting in a head was 100%. We are naturally suspicious of this result, which claims to know the outcome of a future experiment with certainty. The reason our result is suspicious is because we didn't have enough samples Intuition tells us that as we increase the sample size, the above procedure should lead us to an increasingly better estimate for the population mean. There is, in fact, a mathematical result called the <font color='red'>**Law of Large Numbers**</font> which shows that this is true in most cases: it says that <font color='red'>**as we increase the sample size, the sample mean approaches the population (true) mean**</font>.

# Definitions: accuracy and precision

##### The Law of Large Numbers remains vague about how we can define closeness between the sample and population means. In order to quantify this further, we need to differentiate between two measures of "closeness":

>If we know the true value of the quantity we're estimating, then the difference between this true value and our estimate (based on the sample) is called <font color='red'>**accuracy**</font>. The smaller the difference, the better the accuracy. In statistics and machine learning, accuracy is related to another term, the <font color='red'>**bias**</font>.

>As we repeat our experiment, we obtain a different result each time. There is thus a <font color='red'>**variance**</font> in our sample. We can then ask, "what is the <font color='red'>**typical**</font> difference between the different experimental values?" The concept that describes the "closeness" of the results to each other is called <font color='red'>**precision**</font>. Precision measures the <font color='red'>**spread**</font> in our sample obtained by repeating observations. The smaller the spread, the better the precision. In statistics, the precision is related to the <font color='red'>**standard deviation**</font> and its square, the <font color='red'>**variance**</font>, which is also used in machine learning.

>One difference between accuracy and precision is that determining accuracy requires knowledge of the true (population) quantity while precision only has to do with the data. While we always have data, in most cases we do not have access to the true quantity we are trying to estimate.

# The Central Limit Theorem

> Let's consider the case of measuring the length of a string. Each time we measure the length, we are sampling from an underlying population distribution of lengths for the string. The mean of this population is the true length $\mu$ of the string, and there is a spread (measured by the standard deviation $\sigma$ of the distribution) in length values around this true length. Only Nature is aware of the true mean of the string length, we produce estimates for this length from our observations. Each observation/measurement therefore has an uncertainty $\sigma$ associated with it.

> Now let us suppose that we measure the length $N$ times and compute the sample mean of these $N$ lengths. Since the individual measurements are random with a spread $\sigma$, the sample mean, which is a function of these data, is also random and has some spread. <font color='red'>What is the spread/standard deviation of the sample means?</font>.

> According to the Law of Large Numbers, the sample mean is a better (more accurate) estimate of the true length $\mu$. But the sample mean is a random number, so its value also changes. <font color='red'>How, then, do we quantify "more accurate"</font>?

> The Central Limit Theorem informally states that, as the sample size increases,
1.   the **typical** distance between the sample mean and the true mean decreases. That is, **the accuracy of our estimate of the true mean increases**.
2.   the spread in the distribution of sample means decreases. Combined with (1) above, this means that **the precision of our estimates for the true mean increases**.

> More quantitatively, the Central Limit Theorem states that as the number of samples $N$ increases, the distribution of sample means
1.   approaches a Gaussian distribution
2.   the mean of this distribution approaches the true mean
3.   the standard deviation of this distribution approaches $\displaystyle{\sigma\over\sqrt{N}}$



# Illustration
> Let's say that we are measuring the length of a string. We do not know the true value. We will use a ruler whose least count (measurement uncertainty) is 1 mm.</br>
> One measurement results in a value of 800 mm for the string length. We can then say that the true length of the string is likely in the range (800-1, 800+1) or 800±1 mm.</br>
> Now say that we measured the length $N=4$ times, and found that the sample mean of these 4 measurements was 799 mm. The uncertainty associated with this sample mean is $1/\sqrt{4}=1/2=0.5$ mm, therefore we can now say that the true length is likely in the range 799±0.5 or (798.5, 799.5). By increasing the number of observations to 4, we have reduced the spread of possible values to $1/\sqrt{4} = 1/2$ the original spread. The precision with which we know the true mean has therefore increased.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

mu = 798.2 # true length of string in mm
sigma = 1 # uncertainty associated with single measurement in mm

# Let us measure the string length 20 times
Nsamples = 20
lengths = norm.rvs(loc=mu, scale=sigma, size=Nsamples)
# Visualise the distribution of the measured lengths, compare it to the true mean
plt.hist(lengths, histtype='step', label='Measured lengths distribution', density=True)
plt.plot([mu, mu], [0, 1], '--', label='True length')
plt.xlim(796, 800)
plt.legend(loc='best')
```

```python
# Repeat the above sequence 100 times (20 measurements each time),
#   calculate the sample mean each time and plot the distribution of 100 sample means
Nsamples = 20
Ntrials = 100
length_means = norm.rvs(loc=mu, scale=sigma, size=(Nsamples, Ntrials)).mean(axis=1)
# Visualise the distribution of the measured lengths, compare it to the true mean
plt.hist(lengths, histtype='step', label='Measured lengths distribution', density=True)
plt.hist(length_means, histtype='step', label='Sample mean distribution', density=True)
plt.plot([mu, mu], [0, 1], '--', label='True length')
plt.xlim(796, 800)
plt.legend(loc='best')
```

#### As expected, the spread of sample means is much narrower than the spread in individual measurements, AND these sample means are typically much closer to the true mean than the individual measurements.

#### As the total probability (area under the curve) has to be 1, decreasing the spread automatically increases the height of the peak. The interpretation is that computing sample means "focuses" the probability sharply around the population mean.
